```python
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

# Load the grayscale image
gray_image = cv2.imread("grayscale-image.jpg", cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding using a threshold value of 127
threshold_value = 127
thresh, binary_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)

print(binary_image.shape)

# Save the binary image
binary_image_path = "binary-image.jpg"
mpimg.imsave(binary_image_path, binary_image, cmap='gray')

# Display the binary image
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.show()
```
