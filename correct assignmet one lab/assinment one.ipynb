```python
import cv2
from matplotlib import pyplot as plt

# Load the binary image and covert to gray image
binary_image = cv2.imread("binary-image.jpg", cv2.IMREAD_GRAYSCALE)

# convert the image to colorful
color_image = cv2.cvtColor(binary_image, cv2.COLOR_GRAY2BGR)

color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)


# show the shape before and after
print("binary image shape " + str(binary_image.shape))

print("color image shape " +str(color_image.shape))

# show img
plt.imshow(color_image)
plt.axis('off')
plt.show()
```
