```python
import cv2
from matplotlib import pyplot as plt

# read image
img1 = cv2.imread("gabriel-sanchez-jqlXLIj3aS8-unsplash.jpg")
img2 = cv2.imread("eneko-urunuela-lBs-2ZPeSgQ-unsplash.jpg")

# resize image 
y, x, z = img1.shape
img2 = cv2.resize(img2, (x, y))

# add two image (one)
add_imge_one = (img1 + img2) % 256

# add two image (two)
add_image_two = cv2.add(img1, img2)

# show images
plt.subplot(2,2,1), plt.imshow(img1), plt.axis("off"), plt.title("first image")
plt.subplot(2,2,2), plt.imshow(img2), plt.axis("off"), plt.title("second image")
plt.subplot(2,2,3), plt.imshow(add_imge_one), plt.axis("off"), plt.title("first way")
plt.subplot(2,2,4), plt.imshow(add_image_two), plt.axis("off"), plt.title("second way")

plt.show()
```
