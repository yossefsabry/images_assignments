```python
from matplotlib import pyplot as plt
from matplotlib import image as i

# read image
Image_user = i.imread("gabriel-sanchez-jqlXLIj3aS8-unsplash.jpg")

# copy the main image and sperate the channel
im_R = Image_user.copy()
im_R[:, :, (1, 2)] = 0
im_G = Image_user.copy()
im_G[:, :, (0, 2)] = 0
im_B = Image_user.copy()
im_B[:, :, (0, 1)] = 0

# show channel
fig, axes = plt.subplots(1, 3)
axes[0].imshow(im_R)
axes[1].imshow(im_G)
axes[2].imshow(im_B )

plt.show()
```

```python
from PIL import Image
from matplotlib import pyplot as plt

#Create an image object
imageObject = Image.open("gabriel-sanchez-jqlXLIj3aS8-unsplash.jpg")

# split for the image arrays
multImage = imageObject.split()

# draw the histgram for the image
# n, bins, patches = plt.hist(list(multImage[0].getdata()), bins=range(0,256))

# red histgram data
Image_hist=multImage[0].histogram()

# red channel histgram
plt.bar(range(0,256), Image_hist,  color='red')
```
